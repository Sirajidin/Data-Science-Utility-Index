## command: chmod
"set permissions" (aka, assign who can read/write/execute) on a file
chmod permissions file
chmod permission1_permission2_permission3 file

 Therefore, when setting permissions, you are assigning them for "yourself", "your group" and "everyone else" in the world. These users are technically know as:
● Owner
● Group
● World
you will want to assign all three levels of permissions, and not just one user.

chmod owner group world FileName

There are three types of permissions that Linux allows for each file.
● read
● write
● execute

 readable by everyone, and writable by only you,
 COMMAND : OWNER : GROUP : WORLD : PATH
 chmod read & write read read FileName
 chmod 6 4 4 myDoc.txt
 
 Computers like numbers, not words.
 ● 4 – read (r)
 ● 2 – write (w)
 ● 1 – execute (x)
    chmod 400 mydoc.txt – read by owner
    chmod 040 mydoc.txt – read by group
    chmod 004 mydoc.txt – read by anybody (other)
    chmod 200 mydoc.txt – write by owner
    chmod 020 mydoc.txt – write by group
    chmod 002 mydoc.txt – write by anybody
    chmod 100 mydoc.txt – execute by owner
    chmod 010 mydoc.txt – execute by group
    chmod 001 mydoc.txt – execute by anybody
    
Wait! I don't get it... there aren't enough permissions to do what I want!
Good call. You need to add up the numbers to get other types of permissions...

So, try wrapping your head around this!!
7 = 4+2+1 (read/write/execute)
6 = 4+2 (read/write)
5 = 4+1 (read/execute)
4 = 4 (read)
3 = 2+1 (write/execute)
2 = 2 (write)
1 = 1 (execute)
chmod 666 mydoc.txt – read/write by anybody! (the devil loves this one!)
chmod 755 mydoc.txt – rwx for owner, rx for group and rx for the world
chmod 777 mydoc.txt – read, write, execute for all! (may not be the best plan in the world...)

## Use of 'cron'

You can add to and delete entries from you crontab by typing:
crontab -e

Once you have decided when you want a particular command to be run, you add the command itself at the end. Thus, a typical crontab entry will end up looking like this:

30 3 * * 0 $HOME/bkup_script
3:30 AM on Sunday

If you're curious about what's in your crontab file and want to list the jobs you have programmed, type:
crontab -l
If you want to delete your crontab file, type
crontab  -r
